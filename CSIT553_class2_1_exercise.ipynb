## Exercise 3
Which municipality has the largest increasing in population from 2010 to 2017?
The output should be a string.

```python
df["Population Increase"] = df["POP2017"] - df["POP2010"]

largest_increase_municipality = df["Population Increase"].idxmax()

largest_increase_municipality
```
